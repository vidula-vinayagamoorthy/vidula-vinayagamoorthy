## IQR (Interquartile Range)

```python
#Use the interquartile range (IQR) **method** to identify outliers for each feature.

# calculate IQR for first column - Engine rpm

print("IQR calculations for Engine rpm is below:")

# Calculate the quantiles.
q1 = data['Engine rpm'].quantile(0.25)
q3 = data['Engine rpm'].quantile(0.75)

# View the output.
print("The value of Q1 is:", q1)
print("The value of Q3 is:", q3)

# Compute and view the IQR.
iqr = q3 - q1
print("The value of IQR is:",iqr)

# Filtering the outliers. Calculate the upper and lower limits.
lower = q1-1.5*iqr
upper = q3+1.5*iqr

# View the limits.
print("Upper limit: ", upper)
print("Lower limit: ", lower)

# Add column for Engine rpm
# referred to https://stackoverflow.com/questions/21702342/creating-a-new-column-based-on-if-elif-else-condition
#Create a new column (corresponding to each feature) that will indicate (in binary – 0,1)
#if the value of that feature is an outlier as per IQR calculations.
data['Engine rpm IQR'] = np.where(((data['Engine rpm'] > upper) | (data['Engine rpm'] < lower)), 1, 0)

data.head()

```

```python
# calculate IQR for second column - Lub oil pressure

print("IQR calculations for Lub oil pressure is below:")

# Calculate the quantiles.
q1 = data['Lub oil pressure'].quantile(0.25)
q3 = data['Lub oil pressure'].quantile(0.75)

# View the output.
print("The value of Q1 is:", q1)
print("The value of Q3 is:", q3)

# Compute and view the IQR.
iqr = q3 - q1
print("The value of IQR is:",iqr)

# Filtering the outliers. Calculate the upper and lower limits.
lower = q1-1.5*iqr
upper = q3+1.5*iqr

# View the limits.
print("Upper limit: ", upper)
print("Lower limit: ", lower)

# Add column for lub oil pressure
data['Lub oil pressure IQR'] = np.where(((data['Lub oil pressure'] > upper) | (data['Lub oil pressure'] < lower)), 1, 0)

data.head()

```
